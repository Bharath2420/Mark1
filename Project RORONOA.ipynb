# `1-50 `

## Importing the required packages

```python
import requests
import numpy as np
import pandas as pd
import re
from bs4 import BeautifulSoup as bs
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Extracting the desired webpage

```python
anime_base = requests.get('https://myanimelist.net/topanime.php')
```

## Checking for  desireable data 

```python
anime_base.text
```

## Since the data is not clear we use "BeautifulSoup" to extract data out of HTML files

```python
anime_soup = bs(anime_base.text)
```

## Now the data is clear

```python
anime_soup
```

## Extracting the data to the requirements

### Column 1 - Rank

```python
anime_rank=[]
anime = anime_soup.find_all('td',class_='rank ac')
for i in anime:
    data = i.text
    res = re.findall(r'\d+',data)
    A = ''.join(res)
    print(A)
    anime_rank.append(A)   
    
```

### Column 2 - Title

```python
anime_names=[]
anime1 = anime_soup.find_all('a',class_='hoverinfo_trigger fl-l fs14 fw-b')
for i in anime1:
    data = i.text
    print(data)
    anime_names.append(data)
```

```python
anime_lists=[]
anime2 = anime_soup.find_all('div',class_='information di-ib mt4')
for i in anime2:
    data = i.text
    print(data)
    anime_lists.append(data)
```

### Column 3 - Episodes Aired
### Column 4 - Duration
### Column 6 - Members Watched

```python
episodes_list=[]
duration_time=[]
members_viewed=[]
for i in anime2:
    data = i.text
    
    #for episodes
    res = re.findall(r'(\d+).e',data)
    B = ''.join(res)
    #print(B)
    episodes_list.append(B)
    
    #for duration time
    res1 = re.findall(r'\w+.\d+.[-].\w+.\d+',data)
    C = ''.join(res1)
    #print(C)
    duration_time.append(C)
    
    #for  members viewed
    res2 = re.findall(r'\d+.\d+.\d+',data)
    for i in res2:
        new = i.replace(',','')
    D=''.join(new)
    #print(D)
    members_viewed.append(D)
    
    
```

### Column 5 - Ratings

```python
anime_ratings = []
anime3 = anime_soup.find_all('span',class_='text on')
for i in anime3:
    data = i.text
    print(data)
    anime_ratings.append(data)
```

## Creating a dataframe

```python
df_anime = pd.DataFrame()
```

### Assigning each list to a column

```python
df_anime["Rankings"] = anime_rank
df_anime["Title"] = anime_names
df_anime["Episodes"] = episodes_list
df_anime["Duration"] = duration_time
df_anime["Ratings"] = anime_ratings
df_anime["Members"] = members_viewed
```

## Display (DataFrame)

```python
df_anime
```
